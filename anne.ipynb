```python
import cv2
import os

# copying 20 images from pickup folder and converting to grey scale

# Path to your folder with images
input_folder = r"C:\Users\Anne-\datadrevet_programvare\image_processing\data\r7bthvstxw-1\pickup"
output_folder = r"C:\Users\Anne-\datadrevet_programvare\image_processing\data\r7bthvstxw-1\greyscale_pickup"


# Make sure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Loop through only the first 20 image files
for i, filename in enumerate(os.listdir(input_folder)):
    if i >= 20:  # stop after 20 images
        break
    
    img_path = os.path.join(input_folder, filename)
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    cv2.imwrite(os.path.join(output_folder, filename), gray)


```

```python
import numpy as np

# normalizing the images by scaling pixel values to a range [0, 1]

grey_images_folder = r"C:\Users\Anne-\datadrevet_programvare\image_processing\data\r7bthvstxw-1\greyscale_pickup"

size = (128, 128)

before_images = []
normalised_images = []

for f in os.listdir(grey_images_folder):
    img = cv2.imread(os.path.join(grey_images_folder, f), cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, size)  

    before_images.append(img)
    normalised_images.append(img / 255.0)

before_images = np.array(before_images)
normalised_images = np.array(normalised_images)

before_images = np.array(before_images)
normalised_images = np.array(normalised_images)

# print(before_images)
# print(normalised_images[0])
```

```python
# Convert the images into a 2D matrix where each row represents an image and each column represents a pixel value

print(normalised_images.shape)

X = normalised_images.reshape(normalised_images.shape[0], -1)
print("2D matrix shape:", X.shape)

# Compute the covariance matrix of the image data

cov = np.cov(X)
print(cov.shape)
```

```python
from numpy.linalg import eigh

# calculate eigenvectors and eigenvalues for the covariance matrix

eigenvalues, eigenvectors = eigh(cov)

print("number of eigenvectors: ", len(eigenvectors))
print("number of eigenvalues:", len(eigenvalues))
print("eigenvectors: ", eigenvectors)
print("eigenvalues: ", eigenvalues)
```
